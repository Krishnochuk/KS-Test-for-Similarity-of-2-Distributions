# KS test

This performs a test of the distribution G(x) of an observed
random variable against a given distribution F(x). Under the null
hypothesis the two distributions are identical, G(x)=F(x). The
alternative hypothesis can be either 'two-sided' (default), 'less'
or 'greater'. The KS test is only valid for continuous distributions.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
a=stats.norm.rvs(size=1000);   #generate a gaussian random variable a
sns.set_style('whitegrid')
sns.kdeplot(np.array(a),bw=0.5)
plt.show()
```

```python
stats.kstest(a,'norm')
# since the p-value is higher than 5% we accept the null hypothesis
#both the ditributions are same
```

```python
s=np.random.uniform(0,1,size=10000);
sns.set_style('whitegrid')
sns.kdeplot(np.array(s),bw=0.1)
plt.show()
```

```python
stats.kstest(s,'norm')
#since the p-value is zero thus we reject the null hypothesis
#both the distributions are different
```
